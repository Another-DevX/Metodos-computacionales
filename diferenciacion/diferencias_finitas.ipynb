```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Df = lambda f, x, h: (f(x + h) - f(x)) / h
Df_2 = lambda f,x,h: (f(x+h) - f(x-h))/ (2*h)

hs = [10**-n for n in range (1,15)]
print(hs)
x = np.linspace(0, np.pi, 100)
for h in hs:
    plt.plot(x, np.sin(x), label = "sin(x)")
    plt.plot(x, np.cos(x), label="cos(X)")
    plt.plot(x, Df(np.sin, x, h), label=f'Df(x) h={h:.0e}')
    plt.plot(x, Df_2(np.sin,x,h), label = f'Df(x)_2 h={h:.0e}')
    plt.legend()
    plt.grid()
    plt.show()
```
