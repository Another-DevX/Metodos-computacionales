```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
sp.init_printing(use_latex="mathjax")
```

```python
x = sp.Symbol('t')

y = 10*sp.log(x**2 + x + 1)/(10*x**3 - 20 * x ** 2  + x - 2)
y 
```

```python
dy = y.diff(x, 1)
d2y = y.diff(x,2)
display(dy)
f = sp.lambdify(x, y, 'numpy')
df  = sp.lambdify(x, dy, 'numpy')
d2f = sp.lambdify(x,d2y, 'numpy')

x = np.linspace(-1,1,1000)

plt.plot(x, f(x), label="f(x)")
plt.plot(x, df(x), label = "df(x)")
plt.plot(x,d2f(x), label="d2f(x)")
plt.legend()
plt.show()
```
