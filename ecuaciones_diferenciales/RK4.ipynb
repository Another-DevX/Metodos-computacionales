```python
import matplotlib.pyplot as plt
```

```python
def RK4(a, b, alpha, f, N=10):
    h = (b - a) / N
    w0 = alpha
    t = a
    w = [w0]
    t_values = [a]

    for i in range(1, N + 1):
        t = a + i*h
        k1 = h * f(t, w[i - 1])
        k2 = h * f(t + h / 2, w[i - 1] + k1 / 2)
        k3 = h * f(t + h / 2, w[i - 1] + k2 / 2)
        k4 = h * f(t + h, w[i - 1] + k3)
        wi = w[i - 1] + (k1 + 2 * k2 + 2 * k3 + k4) / 6
        w.append(wi)
        t_values.append(t)
    return (t_values, w)
```

```python
f = lambda t, y: y - (t**2) + 1
(a,b) = [0,2]
alpha = 0.5
N = 10


# Solución del método de Euler
t_values, y_values = RK4(a, b, alpha, f, N)


plt.plot(t_values, y_values, '.')
plt.grid()
```
