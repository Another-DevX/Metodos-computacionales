```python
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
def montecarlo(n):
    values = []
    for _ in tqdm(range(n + 1)):
        x = np.random.uniform(-1,1)
        y = np.random.uniform(-1,1)
        if((x ** 2) + (y ** 2) <= 1):
            values.append((x,y))   
            plt.plot(x,y, '.', color='red') 
        else:
            plt.plot(x,y, '.', color='blue') 


    return  4 * (len(values)/n)

montecarlo(100000)

```

```python
def integral_montecarlo(f, M,n, a, b):
    values = []
    for _ in tqdm(range(n + 1)):
        x = np.random.uniform(a,b)
        y = np.random.uniform(0,M)
        if(y <= f(x)):
            values.append((x,y))   
            plt.plot(x,y, '.', color='red') 
        else:   
            plt.plot(x,y, '.', color='blue') 


    return  4 * (len(values)/n)

integral_montecarlo(np.sin, 1, 10000, 0, 2*np.pi)
```
