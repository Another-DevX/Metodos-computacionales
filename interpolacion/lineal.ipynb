```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def interpolacion_lineal(xi,yi):
    if(len(xi) != len(yi)): raise "The vectors must have the same dimension"
    for i in range (1, len(xi)):
        m = (yi[i] - yi[i-1])/(xi[i]-xi[i-1])
        x = np.linspace(xi[i-1], xi[i], 10)
        y = m * (x - xi[i]) + yi[i]

        plt.plot(x, y, '.')
```

```python
xi = np.linspace(0, np.pi, 10)


yi = np.sin(xi)

plt.plot(xi, yi, '+')
interpolacion_lineal(xi,yi)

```
