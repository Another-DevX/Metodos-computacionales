```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
lambd = np.linspace(0.6, 0.8, 1000000)
f_lambda = lambda x : (10000*np.exp(-x)) + (1/x) * (np.exp(-x) - 1) - 5000

plt.plot(lambd, f_lambda(lambd))
plt.xlabel('lambda')
plt.ylabel('f(lambda)')
plt.axhline(0, color='red', linestyle='--')  # Línea recta en y = 0
plt.grid()
plt.xlim(0.680,0.7)

plt.show()
```

```python
x = np.arange(0, np.pi, 0.0001)
plt.plot(x, x, label='x')
plt.plot(x, np.cos(x), label='cos(x)')
plt.grid()
plt.show()
```
