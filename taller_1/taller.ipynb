```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from functools import reduce
```

```python
D0 = np.sqrt(2)/2
N = 100000
n = np.arange(0, N+1)
Dn = np.zeros(N+1)
Dn[0] = D0
for i in range(1,N + 1):
    Dn[i] = np.sqrt(0.5*(1+Dn[i - 1]))
plt.plot(n, Dn)

print(Dn)
```

```python
piN=np.zeros(N+1)
for i in tqdm(range(1,N + 1)):
    piN[i] = 2/np.prod(Dn[0:i+1])
print(piN[-1])
plt.plot(n, piN)

```

```python
N = 100
n0 = 1
n = np.arange(0, N+1)
n_fact = np.zeros(N+1)
n_fact[0] = n0

n_exp = np.zeros(N+1)
n_exp[0] = n0
for i in range(1, N+1):
    n_exp[i] = n_exp[i-1]*2

for i in range(1, N+1):
    n_fact[i] = (i+1)*n_fact[i-1]
print(n_fact)

plt.figure()
plt.title('n!')
plt.semilogy()
plt.plot(n, n_fact)
plt.plot(n, n_exp)

```

```python
f0 = 0
f1 = 1
N = 100
n = np.arange(0, N+1)
fib = np.zeros(N+1)
fib[0] = f0
fib[1] = f1
for i in range(2, N+1):
    fib[i] = fib[i-1] + fib[i-2]
    print(fib[i])

plt.figure()
plt.plot(n, fib, '--')
```
